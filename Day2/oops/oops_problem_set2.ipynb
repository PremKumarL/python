The **flyweight pattern** is used for memory optimization. For programmers new to python programming language, the tendency is to ignore the memory optimization assuming the garbage collector will take care of that. However, when developing large applications with many related objjects, paying a close attention to memory concerns can have a huge payoff.

The idea behind the flyweight pattern is to ensure that the objects that share a state can use the same memory for that shared state. Using the flyweight pattern, we have the shared objects for the list of features associated with a model and then simply reference that model, along with additional features for individual elements.
